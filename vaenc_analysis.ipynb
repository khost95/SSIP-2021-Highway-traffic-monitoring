```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
normal = np.load("/content/drive/MyDrive/aaa/raw/mse_angel_normal.npy")
pause = np.load("/content/drive/MyDrive/aaa/raw/mse_angel_pause.npy")
speedup = np.load("/content/drive/MyDrive/aaa/raw/mse_angel_speedup.npy")
timeflip = np.load("/content/drive/MyDrive/aaa/raw/mse_angel_timeflip.npy")
wrongentry = np.load("/content/drive/MyDrive/aaa/raw/mse_angel_wrongentry.npy")
wrongexit = np.load("/content/drive/MyDrive/aaa/raw/mse_angel_wrongexit.npy")

jam = np.load('/content/drive/MyDrive/aaa/raw/mse_angel_three.npy')
```

```python
fig1, ax1 = plt.subplots(dpi = 150)
ax1.set_title('Raw coordinates')
ax1.set_ylabel('MSE loss')
ax1.boxplot([normal, [], pause, speedup, timeflip, wrongentry, wrongexit], 
            labels = ["normal data", "", "pause", "speed-up", "time-flip", "bad entry", "bad exit"])
plt.show()
```

```python
fig1, ax1 = plt.subplots(dpi = 150)

ax1.set_title('Raw coordinates')
ax1.set_ylabel('MSE loss')
ax1.boxplot([np.log10(normal), [], np.log10(pause), np.log10(speedup), np.log10(timeflip), np.log10(wrongentry), np.log10(wrongexit)], 
            labels = ["normal data", "", "pause", "speed-up", "time-flip", "bad entry", "bad exit"])

ax1.set_yticks(np.arange(-2, 2))
ax1.set_yticklabels(10.0**np.arange(-2, 2))
plt.show()
```

```python
n_bins = 40

fig1, ax1 = plt.subplots(dpi = 150)

ax1.set_xticks(np.arange(-2, 2))
ax1.set_xticklabels(10.0**np.arange(-2, 2))
ax1.set_xlabel('MSE loss')

colors = ['blue', 'red']
labels = ['normal', 'traffic jam']
ax1.hist([np.log10(normal), np.log10(jam)], n_bins, density=True, histtype='bar', color=colors, label=labels)
ax1.legend(prop={'size': 10})
ax1.set_title('Raw coordinates')

plt.show()
```
